```python
import numpy as np
import pandas as pd
import seaborn as sns
```

Load Dataset

```python
# Load the built-in Titanic dataset
df = sns.load_dataset('titanic')

# Look at the first 5 rows to understand the columns
df.head()
```

Adding few columns for demonstrating data issues

```python
df.head()
```

```python
# Add a new column 'passenger_id' with unique values for each passenger
df['passenger_id'] = np.arange(1, len(df)+1)

# Add a constant column with the same value for all rows
df['contant_col'] = 1

#create a dirty version of embark_town column
df['embark_town_dirty'] = df['embark_town'].copy()

# select a few random rows only non-missing values
sample_idx = df['embark_town_dirty'].dropna().sample(6, random_state=42).index

# Apply simple dirty transformations to the selected rows
df.loc[sample_idx [0:2], 'embark_town_dirty'] = df.loc[sample_idx [0:2], 'embark_town_dirty'].str.upper()  # Convert to uppercase
df.loc[sample_idx [2:4], 'embark_town_dirty'] = df.loc[sample_idx [2:4], 'embark_town_dirty'].str.strip().apply(lambda x: f' {x} ')  # Add extra spaces
df.loc[sample_idx [4:6], 'embark_town_dirty'] = 'unknown'  # Replace with unknown

df.head()
```

Missing Values Summary

```python
missing_count = df.isna().sum().sort_values(ascending=False)
missing_percentage = (df.isna().mean()*100).sort_values(ascending=False)

missing_summary = pd.DataFrame({
    
        'missing_count': missing_count, 
        'missing_percentage': missing_percentage
     
     })
print('Missing Values Summary:')
missing_summary
```

*Duplicate Value Check*

```python
duplicate_mask = df.duplicated()
num_duplicates = duplicate_mask.sum()
print(f'Number of duplicate rows: {num_duplicates}')
```

*Data Type Validation*

```python
expected_dtypes = {
    "survived": "int64",
    "pclass": "int64",
    "sex": "category",
    "age": "float64",
    "fare": "float64",
    "embarked": "category",
    "embark_town": "category",
}

for col, expected in expected_dtypes.items():
    if col in df.columns:
        actual_dtype = df[col].dtype
        print(f"{col}: actual = {actual_dtype}, expected = {expected}")
```

```python
df = df.rename(columns={"contant_col": "const_columns"})
df.columns
```

**Constant and Quasi-Constant Columns**

```python
n_rows = len(df)
print('Number of Rows:', n_rows)
nunique = df.nunique()
constant_col = nunique[nunique==1].index.tolist() 
print('Constant Columns:', constant_col)

quasi_constant_cols = []

for col in df.columns:
    top_freq = df[col].value_counts(normalize=True, dropna=False).values[0]
    if top_freq > 0.95 and col not in constant_col:
        quasi_constant_cols.append(col)

print('Quasi-Constant Columns:', quasi_constant_cols)
```

```python
nunique = df.nunique()
print('Unique Values per Column:', nunique)
```

**Column With All Unique Values**

```python
n_rows = len(df)

column_with_all_unique = []

for col in df.columns:
    unique_count = df[col].nunique()
    if unique_count == n_rows:
        column_with_all_unique.append(col)
print(f"Columns with all unique values: {column_with_all_unique}")
```
