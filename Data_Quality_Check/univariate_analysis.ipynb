```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
pd.set_option('display.max_columns', 50)
sns.set_style('whitegrid')
```

##### 2. Load dataset

```python
# Load the Titanic dataset
data = sns.load_dataset('titanic')

# Look at the first 5 rows
display(data.head())
```

```python
# Identify numerical and categorical columns
numerical_cols = ['age', 'fare', 'sibsp', 'parch']  # Explicitly specify numerical columns
categorical_cols = ['survived', 'pclass', 'sex', 'embarked', 'class', 'who', 'adult_male', 'deck','embark_town', 'alone']

print("Categorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)
```

##### 3. Numerical Features Analysis

```python
data[numerical_cols].head()

```

3.1 Summary Statistics

```python
data[numerical_cols].describe()
```

3.2 Mean

```python
# Mean

print('Numerical Columns Mean:')
print(data[numerical_cols].mean())

print('-'*50)

print('Numerical Columns Median:')
print(data[numerical_cols].median())

print('Numerical Columns Mode:')
print(data[numerical_cols].mode().iloc[0])
```

3.3 Varience and Standard Deviation

```python
#Variance
print('Numerical Columns Variance:')
print(data[numerical_cols].var())

# Standard Deviation
print('Numerical Columns Standard Deviation:')
print(data[numerical_cols].std())
```

3.4 Skewness and Kurtosis

```python
print('\n Skewness of Numerical Data')
print(data[numerical_cols].skew())
print('\n Kurtosis of Numerical Data')
print(data[numerical_cols].kurt())
```

3.5 Histograms

```python
plt.figure(figsize=(5,3))
sns.histplot(data['age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')   
plt.ylabel('Frequency')
plt.show()
```

```python
# Numerical Columns Histograms
for col in numerical_cols:
    plt.figure(figsize=(5,3))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f'Distribution of {col.capitalize()}')
    plt.xlabel(col.capitalize())   
    plt.ylabel('Frequency')
    plt.show()
```
